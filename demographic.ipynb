```python
import pandas as pd
pd.set_option('display.max_columns', 48)
```

```python
d106 = pd.read_excel('Data/demographic/age_gender106.xlsx', header=1)
```

```python
d106.shape
```

```python
col = d106.columns.tolist()
```

```python
co = []
for c in col:
    if '合計' in c or '區' in c or '性別' in c or '總' in c:
        co.append(c)
```

```python
co
```

```python
data106 = d106.reindex(columns=co).rename(columns={'區 域 別': 'region', '性別 ': 'gender', '總計':'total'})
```

```python
gender_dict = {'計':'sum', '男':'M', '女':'F'}
```

```python
data106.gender = data106.gender.map(gender_dict)
```

```python
data106.head(20)
```

```python
data106.to_csv('Data/demographic/age_gender17.csv', index=True)
```

```python
data106_piv = data106.pivot_table(index='region', columns='gender')
```

```python
data106_piv.columns = data106_piv.columns.to_series().str.join('_')
```

```python
data106_piv.columns = data106_piv.columns.str.replace('合計_', '')
```

```python
data106_piv.columns = 'y17_' + data106_piv.columns
```

```python
data106_piv.columns = data106_piv.columns.str.replace('歲', '')
```

```python
data106_piv.columns = data106_piv.columns.str.replace('~', '')
```
